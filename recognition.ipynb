```python
import cv2
import os
import numpy as np
import face_recognition
```

```python
path='images'
```

```python
mylist=os.listdir(path)
mylist
```

```python
imgs=[]
classname=[]
for img in mylist:
    imgpath=os.path.join(path,img)
    print(imgpath)
    current_img=cv2.imread(imgpath)
    print(current_img)
    imgs.append(current_img)
    classname.append(img[:-5])
print(len(imgs))
print(classname)
```

```python
img=imgs[0]
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
face_in_frame=face_recognition.face_locations(img)
face_in_frame
```
